```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the two datasets
# Merge them using the common Patient_ID
# Display the first 10 rows in a heatmap-like style

```python
patient_info = pd.read_csv('Patient_Info.csv')
diagnosis_info = pd.read_csv('Diagnosis_Info.csv')

merged_df = pd.merge(patient_info, diagnosis_info, on='Patient_ID')
#merged_df.head(10).style.set_table_styles([{'selector': 'th', 'props': [('background-color', '#4A90E2'), ('color', 'white')]}]).set_properties(**{'border': '1px solid black', 'padding': '8px'}).background_gradient(cmap='Blues')

sns.set(style='whitegrid')
plt.figure(figsize=(6, 4))
sns.countplot(data=merged_df, x='Risk_Level', palette='Blues')
plt.title('Number of Patients by Risk Level')
plt.xlabel('Risk Level')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(data=merged_df, x='Risk_Level', y='Blood sugar', palette='Blues')
plt.title('Blood Sugar Distribution Across Risk Levels')
plt.xlabel('Risk Level')
plt.ylabel('Blood Sugar (mg/dL)')
plt.tight_layout()
plt.show()
```
